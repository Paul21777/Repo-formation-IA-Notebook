# Importation des librairies

```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

# Chargement et préparation et conversiondes données (Iris Dataset)

```python
iris = load_iris()
X = iris.data[:, 2:]  # Utiliser seulement les caractéristiques "petal length" et "petal width"
y = iris.target
target_names = iris.target_names
```

# S'assurer graphiquement de la bonne répartitions des données

```python
# Création du plot
plt.figure(figsize=(8, 6))

# Parcours des classes
for target in range(len(target_names)):
    # Sélection des exemples de la classe courante
    X_class = X[y == target]
    
    # Plot des exemples de la classe courante
    plt.scatter(X_class[:, 0], X_class[:, 1], label=target_names[target])

# Paramètres du plot
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Distribution des données de l'Iris Dataset")
plt.legend()

# Affichage du plot
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) 
```

```python
grid_parameters = {'max_depth': range(3,10), 'random_state' : [42]}
```

```python
grid_search = GridSearchCV(DecisionTreeClassifier(), grid_parameters, cv=5, n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
print('Best parameters:', grid_search.best_params_)
print('Best score:', grid_search.best_score_)
```

```python
grid_search.fit(X_test, y_test)
print('Best parameters:', grid_search.best_params_)
print('Best score:', grid_search.best_score_)
```
